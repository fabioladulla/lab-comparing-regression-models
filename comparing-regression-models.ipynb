```python
def split_numerical(df, max_unique_values=36):
    num_df = df.select_dtypes(include=[np.number])
    continuous_columns = [col for col in num_df.columns if num_df[col].nunique() > max_unique_values]
    discrete_columns = [col for col in num_df.columns if num_df[col].nunique() <= max_unique_values]
    
    return df[continuous_columns], df[discrete_columns]
```

```python
# Define a function to remove the outliers of a numerical continuous column depending if a value is bigger or smaller than a given amount of standard deviations of the mean (thr=3).
def remove_outliers(df, column, thr=3):
    """
    Remove outliers from a dataframe column based on the threshold value for standard deviations
    
    Parameters:
    df (pd.DataFrame): Input DataFrame
    column (str): Column name
    thr (int): Threshold value for standard deviations. Default is 3.
    
    Returns:
    pd.DataFrame: DataFrame with outliers removed
    """
    mean = df[column].mean()
    std = df[column].std()
    
    lower_limit = mean - thr * std
    upper_limit = mean + thr * std

    filtered_df = df[(df[column] >= lower_limit) & (df[column] <= upper_limit)]
    
    return filtered_df
```

```python
# Use the previous function to remove the outliers of continuous data and to generate a continuous_clean_df.
continuous_clean_df = continuous_df.copy()

for column in continuous_clean_df.columns:
    continuous_clean_df = remove_outliers(continuous_clean_df, column, thr=3)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def error_metrics(y, y_pred):
    return pd.DataFrame({
        'Error Metric': ['Mean Squared Error', 'Mean Absolute Error', 'R-squared'],
        'Value': [mean_squared_error(y, y_pred), 
                  mean_absolute_error(y, y_pred), 
                  r2_score(y, y_pred)]
    })

# Get error metrics for the training set
train_errors = error_metrics(y_train, y_train_pred)
print(train_errors)

# Get error metrics for the test set
test_errors = error_metrics(y_test, y_test_pred)
print(test_errors)
```

```python
def train_models(models, X_train, y_train):

    for model in models:
        model.fit(X_train, y_train)
    
    return models
```
